```python
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams.update({'figure.figsize': (12, 8),
                        'figure.dpi' : 80,
                        'axes.grid': True})
```

```python
df = pd.read_csv('AirPassengers.csv') #日期标签名
df.head()
```

```python
df.set_index('Month', inplace=True)
df.index = pd.to_datetime(df.index)
df.dropna(inplace=True)
```

```python
result = seasonal_decompose(df['#Passengers'], model='multiplicable', period = 12)
```

```python
plt.figure(figsize=(15, 2), dpi=300)
plt.plot(df.iloc[:, 0])
plt.xlabel('time series', fontsize=12)
plt.ylabel('theta', fontsize=14)
# plt.savefig('mu.jpg', bbox_inches='tight')
plt.show()

plt.figure(figsize=(15, 2), dpi=300)
plt.plot(result.trend)
# plt.plot(yy, linestyle='--', linewidth=3)
plt.xlabel('time series', fontsize=12)
plt.ylabel('Trend', fontsize=14)
# plt.savefig('mu-trend.jpg', bbox_inches='tight')
plt.show()

plt.figure(figsize=(15, 2), dpi=300)
plt.plot(result.seasonal)
plt.xlabel('time series', fontsize=12)
plt.ylabel('Periodicity', fontsize=14)
# plt.savefig('mu-period.jpg', bbox_inches='tight')
plt.show()

plt.figure(figsize=(15, 2), dpi=300)
plt.plot(result.resid, marker='o', linestyle='none')
plt.xlabel('time series', fontsize=12)
plt.ylabel('Residual', fontsize=14)
# plt.savefig('mu-res.jpg', bbox_inches='tight')
plt.show()
```
